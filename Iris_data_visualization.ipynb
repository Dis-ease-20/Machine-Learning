**Import** **Library**





```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Dataset**

```python
data_df = sns.load_dataset('iris')
```

```python
data_df
```

**Check the unique species in dataset**

```python
data_df.species.unique()
```

**Scatter Plot**

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,8))
plt.title("Sepal Dimenaions")
sns.scatterplot(x = data_df.sepal_length, y = data_df.sepal_width, hue = data_df.species,s =100)
```

**Line Plot**

```python
plt.figure(figsize=(10,8))
plt.title("Sepal Dimenaions")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.plot(data_df.sepal_length, data_df.sepal_width)
```

**Bar graph**

```python
plt.figure(figsize=(10,8))
plt.title("Petal length of Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.bar(data_df.species, data_df.petal_length, color="orange");
```

**Histogram Plot**

```python
setosa_df = data_df[data_df.species == 'setosa']
versicolor_df = data_df[data_df.species == 'versicolor']
virginica_df = data_df[data_df.species == 'virginica']
```

```python
import numpy as np
plt.figure(figsize=(14,8))
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```
